```python
#!/bin/bash
# use this script to demonstrate if...then..else and the test command
# run this in a linux environment
# example:
# if-then is the name of the script file
# [lisa@localhost scripts]$ chmod +x ./if-then
# [lisa@localhost scripts]$ ./if-then /home/lisa/Downloads/

if [ -f $1 ]
then
	echo $1 is a file
elif [ -d $1 ]
then
	echo $1 is a directory
else
	echo I do not know what \$1 is
fi
exit 0
```

```python
#!/bin/bash
# use this to demonstrate logical and && and logical or || 
# &&: it will exectue the second part only if the first part is true
[ -z $1 ] && echo no argument provided
# ||: it wil execute the second part only if the first part is not true
ping -c 1 10.0.0.20 2>/dev/null || echo node is not available
```

```python
#!/bin/bash
# use this script to demonstrate for loop
for ((COUNTER=100; COUNTER>1; COUNTER--)); do
    echo $COUNTER
done
exit 0
```

```python
#!/bin/bash
# use this script to demonstarte one-liner for loop
# notice how the range is defined: {100..104}
# ping command result doesn't count, we only take the exit status here
for i in {100..104}; do ping -c 1 192.168.4.$i >/dev/null && echo 192.168.4.$1 is up; done
```

```python
#!/bin/bash
# use this script to demonstrate while loop
# usage: monitor <processname>
# grep <processname> from the output of ps command, excludes the line containing grep command
while ps aux | grep $1 | grep -v grep > /dev/tty11
do
    sleep 5
done
clear
echo your process has stopped
logger $1 is no longner present
# mail to the user root, with a subject, and no message body
# message body is skippped by passing a line with only a dot,  i.e. < .
mail -s "process $1 has stopped" root < .
```

```python
#!/bin/bash
# use this script to demonstrate until loop
# usage: monitor user login
until users | grep $1 > /dev/null
do
    echo $1 is not logged in yet
    sleep 5
done
echo $1 has just logged in
# mail to the user root, with a subject, and no message body
# message body is skippped by passing a line with only a dot,  i.e. < .
mail -s "$1 has just logged in" root < .
```
